```python
# 1. Připojení Drive a klonování projektu
from google.colab import drive
import os
import sys

drive.mount('/content/drive')

REPO_URL = "https://github.com/Brkoszov/Omnivoice.git"
PROJECT_DIR = "/content/Omnivoice"
DRIVE_MODELS_PATH = "/content/drive/MyDrive/OmniVoice_Models"

if not os.path.exists(PROJECT_DIR):
    !git clone {REPO_URL} {PROJECT_DIR}
    print("✅ Projekt stažen.")
else:
    print("✅ Projekt již existuje, aktualizuji z GitHubu...")
    %cd {PROJECT_DIR}
    !git pull

if os.getcwd() != PROJECT_DIR:
    %cd {PROJECT_DIR}

MODELS_TARGET_DIR = os.path.join(PROJECT_DIR, "models")
if os.path.exists(MODELS_TARGET_DIR):
    !rm -rf {MODELS_TARGET_DIR}
!ln -s {DRIVE_MODELS_PATH} {MODELS_TARGET_DIR}
print(f"✅ Modely propojeny do {MODELS_TARGET_DIR}")
```

```python
# 2. Stabilní instalace závislostí (Opravená verze)
import os

PROJECT_DIR = "/content/Omnivoice"
REQ_PATH = os.path.join(PROJECT_DIR, "requirements.txt")

print("🧹 Čistím prostředí od nekompatibilních verzí...")
!pip uninstall -y numpy torch torchvision torchaudio

print("📦 Instaluji kompatibilní NumPy 1.26.4...")
!pip install numpy==1.26.4

print("🔥 Instaluji Torch stack (toto může trvat pár minut, neruš to!)...")
!pip install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu121

if os.path.exists(REQ_PATH):
    print(f"🛠️ Instaluji zbytek z {REQ_PATH}...")
    !pip install -r {REQ_PATH}
else:
    print(f"❌ CHYBA: Soubor {REQ_PATH} nebyl nalezen!")

!apt-get install -y ffmpeg
print("\n✅ Všechny závislosti byly nainstalovány.")
print("\n⚠️ DŮLEŽITÉ: Nyní klikni na 'Runtime' -> 'Restart session' a pak pokračuj buňkou č. 3.")
```

```python
# 3. Automatická úprava app.py pro veřejný odkaz (share=True)
import os
import re

PROJECT_DIR = "/content/Omnivoice"
app_py_path = os.path.join(PROJECT_DIR, 'app.py')

if os.path.exists(app_py_path):
    with open(app_py_path, 'r') as f:
        content = f.read()

    if 'share=True' not in content:
        content = re.sub(r'launch\((.*?)\)', r'launch(share=True, \1)', content)
        if 'launch(' not in content:
            content = content.replace('launch()', 'launch(share=True)')
        
        with open(app_py_path, 'w') as f:
            f.write(content)
        print("✅ app.py byl upraven pro veřejný odkaz.")
    else:
        print("✅ app.py již obsahuje share=True.")
else:
    print("❌ Soubor app.py nebyl nalezen!")
```

```python
# 4. Spuštění aplikace
import sys
import os

PROJECT_DIR = "/content/Omnivoice"
if PROJECT_DIR not in sys.path:
    sys.path.append(PROJECT_DIR)

print("🚀 Spouštím OmniVoice... Počkej na odkaz .gradio.live")
!python {os.path.join(PROJECT_DIR, 'app.py')}" 
```

```python
# 1. Připojení Drive a klonování projektu
from google.colab import drive
import os
import sys

drive.mount('/content/drive')

REPO_URL = "https://github.com/Brkoszov/Omnivoice.git"
PROJECT_DIR = "/content/Omnivoice"
DRIVE_MODELS_PATH = "/content/drive/MyDrive/OmniVoice_Models"

if not os.path.exists(PROJECT_DIR):
    !git clone {REPO_URL} {PROJECT_DIR}
    print("✅ Projekt stažen.")
else:
    print("✅ Projekt již existuje, aktualizuji z GitHubu...")
    %cd {PROJECT_DIR}
    !git pull

if not os.getcwd() == PROJECT_DIR:
    %cd {PROJECT_DIR}

MODELS_TARGET_DIR = os.path.join(PROJECT_DIR, "models")
if os.path.exists(MODELS_TARGET_DIR):
    !rm -rf {MODELS_TARGET_DIR}
!ln -s {DRIVE_MODELS_PATH} {MODELS_TARGET_DIR}
print(f"✅ Modely propojeny do {MODELS_TARGET_DIR}")
```
